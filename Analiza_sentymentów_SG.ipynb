```python
# Import biblioteki pandas
import pandas as pd

# Wczytanie pliku csv
df = pd.read_csv('HotelReviews.csv', sep=';')

# Wyświetlenie tabeli z pliku csv
df.head()
```

```python
# Połączenie kolumn 'Positive_Review' i 'Negative_Review' do jednej kolumny 'Review'
df['Review'] = df['Negative_Review'] + ' ' + df['Positive_Review']

# Wyświetlenie pierwszych 5 wierszy, aby sprawdzić, czy operacja się powiodła
df.head()
```

```python
# Zakodowanie ogólnej oceny hotelu w formie binarnej
df['Review_Score'] = df['Reviewer_Score'].apply(lambda x: 0 if x >= 5 else 1)

# Obliczenie liczby pozytywnych i negatywnych recenzji
positive_reviews = df[df['Review_Score'] == 0].shape[0]
negative_reviews = df[df['Review_Score'] == 1].shape[0]

print(f'Liczba pozytywnych recenzji: {positive_reviews}')
print(f'Liczba negatywnych recenzji: {negative_reviews}')
```

```python
# Usunięcie wyrażeń 'No Positive' i 'No Negative' z recenzji
df['Review'] = df['Review'].replace('No Positive', '')
df['Review'] = df['Review'].replace('No Negative', '')

# Wyświetlenie pierwszych 5 wierszy, aby sprawdzić, czy operacja się powiodła
df.head()
```

```python
# Zamiana wszystkich wielkich liter na małe
df['Review'] = df['Review'].str.lower()

# Wyświetlenie pierwszych 5 wierszy, aby sprawdzić, czy operacja się powiodła
df.head()
```

```python
# Wgranie biblioteki nltk
import nltk
# Pobranie słów kluczowych 'stopwords'
from nltk.corpus import stopwords
# Pobranie funkcji word_tokenize, która służy do dzielenia tekstu na poszczególne słowa (tokeny)
from nltk.tokenize import word_tokenize

# Pobranie listy angielskich przerywników
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Usunięcie przerywników i jednoliterowych słów
df['Review'] = df['Review'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word not in stop_words and len(word) > 1]))

# Wyświetlenie pierwszych 5 wierszy, aby sprawdzić, czy operacja się powiodła
df.head()
```

```python
# Import klasy WordNetLemmatizer
from nltk.stem import WordNetLemmatizer

# Wgranie zasobu WordNet, który jest używany przez WordNetLemmatizer do lematyzacji słów.
# Lematyzacja to proces sprowadzania słowa do jego podstawowej, kanonicznej formy, zwanej lematem. 
nltk.download('wordnet')

# Inicjalizacja lematyzatora
lemmatizer = WordNetLemmatizer()

# Lematyzacja słów
df['Review'] = df['Review'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(x)]))

# Wyświetlenie pierwszych 5 wierszy, aby sprawdzić, czy operacja się powiodła
df.head()
```

```python
# Import klasy SentimentIntensityAnalyzer z modułu vader w bibliotece nltk. 
# Analizuje i ocenia emocjonalny wydźwięk tekstu, np. czy tekst wyraża pozytywne, negatywne lub neutralne uczucia.
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Wgranie zasobu vader_lexicon, który jest używany przez SentimentIntensityAnalyzer do analizy sentymentów
nltk.download('vader_lexicon')

# Inicjalizacja analizatora sentymentów
sia = SentimentIntensityAnalyzer()

# Dodanie kolumny z wynikami analizy sentymentów
df['Sentiment'] = df['Review'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Wybranie recenzji, które mają więcej niż 5 słów
df = df[df['Review'].apply(lambda x: len(word_tokenize(x)) > 5)]

# Wyświetlenie 10 najlepszych recenzji
print('10 najlepszych recenzji:')
print(df.nlargest(10, 'Sentiment')['Review'])

# Wyświetlenie 10 najgorszych recenzji
print('\n10 najgorszych recenzji:')
print(df.nsmallest(10, 'Sentiment')['Review'])
```

```python
# Import funkcji train_test_split z modułu model_selection w bibliotece scikit-learn.
# Dzieli dostępne dane na dwie części, aby móc trenować model na jednym zbiorze i oceniać jego skuteczność na drugim zbiorze
from sklearn.model_selection import train_test_split

# Import klasy LogisticRegression z modułu linear_model w bibliotece scikit-learn.
# Dostarcza funkcje i metody do budowy i dopasowania modelu regresji logistycznej.
from sklearn.linear_model import LogisticRegression

# Import funkcji accuracy_score z modułu metrics w bibliotece scikit-learn.
# Oblicza dokładność klasyfikacji.
from sklearn.metrics import accuracy_score

# Podział zbioru danych na dane uczące i testowe
X_train, X_test, y_train, y_test = train_test_split(df[['Sentiment']], df['Review_Score'], test_size=0.2, random_state=42)

# Inicjalizacja klasyfikatora
clf = LogisticRegression()

# Trenowanie klasyfikatora
clf.fit(X_train, y_train)

# Przewidywanie etykiet dla danych testowych
y_pred = clf.predict(X_test)

# Obliczenie i wyświetlenie dokładności klasyfikacji
accuracy = accuracy_score(y_test, y_pred)
print(f'Dokładność klasyfikacji: {accuracy}')
```
